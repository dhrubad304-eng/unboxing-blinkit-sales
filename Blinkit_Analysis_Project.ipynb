## **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## **Import Raw Data**

```python
df = pd.read_csv(r"blinkit_data.csv")
```

## **Data Cleaning**

```python
print(df["Item Fat Content"].unique())
```

```python
df["Item Fat Content"] = df["Item Fat Content"].replace({"LF" : "Low Fat",
    "low fat" : "Low Fat",
    "reg" : "Regular" })             
```

```python
print(df['Item Fat Content'].unique())
```

## **BUSINESS REQUIRMENTS**

### **KPI'S REQUIRMENTS**

```python
#Total Sales
total_sales = df["Sales"].sum()
print(f"The toal sales is: {total_sales}")

#Average Sales
avg_sales = df["Sales"].mean()
print(f"The average sales is: {avg_sales}")

#Number of item sold
no_of_item_sold = df["Sales"].count()
print(f"The Number of item sold are: {no_of_item_sold}")

#Average Rating
avg_rating = df["Rating"].mean()
print(f"The average Rating is: {avg_rating}")
```

## **VISULIZATION FOR BUSINESS REQUIRMENTS**

#### **Total Sales By Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat,
        labels=sales_by_fat.index,
        autopct='%.1f%%',
        startangle=90)

plt.title("Total Sales by Fat Content")
plt.axis("equal")  
plt.show()
```

#### **Total Sales By Item Type** 

```python
sales_by_type = df.groupby("Item Type")['Sales'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(data=sales_by_type, x='Item Type', y='Sales')

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet for Total Sales** 

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().reset_index()

fig = px.bar(grouped,
             x='Outlet Location Type',
             y='Sales',
             color='Item Fat Content',
             barmode='group')

fig.show()
```

#### **Total Sales By Outlet Establishment**  

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(10,5))
fig = px.line(
    x=sales_by_year.index,
    y=sales_by_year.values,
    markers=True,
    title="Total Sales by Outlet Establishment Year",
    labels={'x': 'Outlet Establishment Year', 'y': 'Total Sales'}
)

fig.show()

```

#### **Sales By Outlet Size**  

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels = sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

#### **Sales By Outlet Location**  

```python
sales_by_location = df.groupby("Outlet Location Type")['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(8,3))
ax = sns.barplot(x='Sales',y='Outlet Location Type',data=sales_by_location)

plt.title('Total Sales By Outlet Location')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```
